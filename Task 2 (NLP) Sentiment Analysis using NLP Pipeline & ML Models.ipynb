```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
# Lowercasing
def to_lower(text):
    return text.lower()
    
```

```python
# Remove URLs & Special Characters
def remove_urls(text):
    return re.sub(r'http\S+|www\S+|https\S+', '', text)

def remove_special_chars(text):
    return re.sub(r'[^a-zA-Z\s]', '', text)
```

```python
#removing punctuation
def remove_punctuation(text):
    return text.translate(str.maketrans('', '', string.punctuation))
```

```python
# Tokenization
def tokenize(text):
    return word_tokenize(text)
```

```python
# Stemming
def remove_stopwords(tokens):
    return [word for word in tokens if word not in stop_words]
```

```python
df['text'] = df['text'].fillna("")
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

def lemmatize(tokens):
    return [lemmatizer.lemmatize(word) for word in tokens]
```

```python
def preprocess(text):
    if not isinstance(text, str):
        return ""
    
    text = to_lower(text)
    text = remove_urls(text)
    text = remove_punctuation(text)
    text = remove_special_chars(text)
    
    tokens = tokenize(text)
    tokens = remove_stopwords(tokens)
    tokens = lemmatize(tokens)
    
    return " ".join(tokens)
```

```python

import nltk
nltk.download('punkt_tab')
```

```python
df['clean_text'] = df['text'].apply(preprocess)
```

 Insights & Analysis

Decision Tree Performance
- Moderate accuracy (~69.6%) on test data.  
- Precision and recall are roughly equal, meaning the model is balanced across predicting positive, negative, and neutral sentiments.  
- F1 score (~0.694) confirms that overall performance is consistent but not perfect.  

 Data Imbalance Effect
- Airline sentiment datasets often have more neutral/negative than positive tweets, which can influence metrics.  
- The higher accuracy on the full dataset (~0.848) may be biased toward dominant classes.  

 Feature Representation
- TF-IDF captures the most frequent terms, but contextual meaning is lost, so the model struggles on subtle sentiment expressions.  
- Using **Word2Vec**, **BERT embeddings**, or **n-grams** might improve performance.  

 Decision Tree Limitations
- Trees can overfit on training data.  
- Might benefit from **Random Forest** or **Gradient Boosting** for better generalization.  

Model Overall Comparison
- **Logistic Regression** performed best overall because it handles text features effectively.  
- **Naive Bayes** is fast but assumes independence between words, which is not always realistic.  
- **Decision Tree** gives moderate accuracy but tends to overfit on text data.  
- **TF-IDF** gave better results than Bag of Words as it considers word importance.
