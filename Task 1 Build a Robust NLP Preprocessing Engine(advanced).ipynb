1."Love" vs "love" in NLP

In NLP, “Love” and “love” can be treated as different words (tokens) in case-sensitive models because of the difference in capital letters. “Love” may represent a proper noun or the beginning of a sentence, while “love” is usually a common word.

However, in many NLP applications, all text is converted to lowercase. In that case, both “Love” and “love” are treated as the same word, which helps reduce vocabulary size, improve consistency, and simplify processing.

2.What happens if stopwords are not removed?

Stopwords are very common words like the, is, in, on.
If we do not remove them, the text will contain many unnecessary words, which:
-Makes processing slower
-Adds noise to the data
-Can reduce the performance of models

However, sometimes these words are important for meaning

3.Provide two real-world scenarios where removing stopwords can be harmful.

-There are situations where stopwords are important:

-Sentiment Analysis: Words like not, no, never change meaning. Example: “not good” becomes “good” if “not” is removed, which is wrong.
-Question Answering / Search: Removing stopwords can make sentences unclear. Example: “What is the capital of India?” becomes “capital India”, losing proper meaning.

4.Explain the difference between stemming and lemmatization.

-Both are used to reduce words to their base form.

Stemming:
It simply cuts off word endings(suffix) without caring about meaning.
The result may not be a real word.
Example :studies → studi

Lemmatization:
It converts the word into its correct base form using meaning and grammar.
The result is always a proper word.
Example: studies → study

Stemming is faster but less accurate, while lemmatization is slower but gives better results.

```python
# Task 2: Build Advanced Preprocessing Function

import re

def preprocess_text(text):
    # Lowercase
    text = text.lower()
    
    # Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)
    
    # Remove emails
    text = re.sub(r'\S+@\S+', '', text)
    
    # Remove numbers
    text = re.sub(r'\d+', '', text)
    
    # Handle repeated characters (soooo → so)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    
    # Remove punctuation and emojis
    text = re.sub(r'[^\w\s]', '', text)
    
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    # Remove short words (≤2), keep "no", "not"
    tokens = text.split()
    filtered_tokens = [
        word for word in tokens
        if len(word) > 2 or word in ["no", "not"]
    ]
    
    cleaned_sentence = " ".join(filtered_tokens)
    
    return filtered_tokens, cleaned_sentence
```

```python
# Task 3: Stress Testing

sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

# code:
for i, text in enumerate(sentences, 1):
    tokens, cleaned = preprocess_text(text)
    
    print(f"{i}.")
    print("Original Text:")
    print(text)
    print("Cleaned Tokens:")
    print(tokens)
    print("Cleaned Sentence:")
    print(cleaned)
    print("-" * 40)
```

```python
# Task 4: Token Analytics

total_tokens = len(tokens)
unique_tokens = len(set(tokens))

avg_len = sum(len(t) for t in tokens) / total_tokens

print("Total Tokens:", total_tokens)
print("Unique Tokens:", unique_tokens)
print("Average Length:", round(avg_len, 2))
```

```python
# Task 5: Frequency Analysis

from collections import Counter

all_tokens = []

# Collect all tokens
for text in sentences:
    tokens, _ = preprocess_text(text)
    all_tokens.extend(tokens)

# Apply Counter
word_count = Counter(all_tokens)

# Top 10 most frequent words
print("Top 10 Most Frequent Words:")
for word, freq in word_count.most_common(10):
    print(word, ":", freq)

# Top 5 least frequent words
print("\nTop 5 Least Frequent Words:")
least_common = word_count.most_common()[::-1]  # reverse list

for word, freq in least_common[:5]:
    print(word, ":", freq)
```

```python
# Task 6: Build Full Pipeline 

def full_pipeline(text_list):

    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
result = full_pipeline(sentences)
print(result)

    
```
